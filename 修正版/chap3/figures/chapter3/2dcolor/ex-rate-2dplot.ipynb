```python

import math
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt




def excitation_rate_2color(f_width,frep ,P_760, P_894):
    #物理定数など
    e0 = 8.854 * (10**-12)  # Fm**-1
    c = 2.998*(10**8) # m/s 
    h_ = 1.054*10**(-34) #J*s
    e = 1.602 * 10**(-19)
    bracket = 4.1*10**(-22)
    r = 2*math.pi*2.18*1000000
    
    n = f_width/frep
    delta = 0
    
    I_760 = P_760 /(((0.25*10**(-3))**2)*math.pi)
    I_894 = P_894 / (((0.25*10**(-3))**2)*math.pi)
    E_760 = (2*I_760/(n*c*e0))**0.5
    E_894 = (2*I_894/(n*c*e0))**0.5
    
    for i in range(int(n/2)):
        delta += ((f_width/2+2*10**9) + frep*i)**(-1) 
    for i in range(int(n/2)):
        delta += ((f_width/2+2*10**9) - frep*(i+1))**(-1)

    W = e**2 * E_760 * E_894 * bracket * delta * 0.5 / (h_**2)

    r_N = W**2/r
    
    return r_N
    
n = excitation_rate_2color(5*10**12,1.6*10**9,1,0.01)

print(n)
```
